```python
import pandas as pd
```

```python
column_names=['user_id','item_id','rating','timestamp']
df=pd.read_csv('u.data',sep='\t',names=column_names)
```

```python
df.head()
```

```python
len(df) #number of enteries in the dataset
```

```python
m_cols=['item_id','title']
movie_titles = pd.read_csv('u.item', sep='|', names=m_cols, usecols=range(2), encoding='latin1')
movie_titles.head()
```

```python
len(movie_titles)
```

```python
df=pd.merge(df,movie_titles,on='item_id')
df.head()
#user id shows reviewer UID
#item id shows movie ID
#title and rating shows title and rating of respective movies
```

```python
#creating a pivot table
#each row is user and column is movie name
#table contains rating values 
moviepivot=df.pivot_table(index='user_id',columns='title',values='rating')
moviepivot.head()
```

```python
type(moviepivot)
```

```python
#example of starwars movie
starwars_UR=moviepivot['Star Wars (1977)']
starwars_UR.head()
```

```python
#finding movies simialr to star wars
similar=moviepivot.corrwith(starwars_UR)
```

```python
similar
#higher the number, better the correlation
```

```python
type(similar)
```

```python
#showing the table in a cleaner formal
correlation=pd.DataFrame(similar,columns=['Correlation'])
correlation.dropna(inplace=True)
correlation.head()
```

```python
#finding movies most simialar to starwars
#results are a little inaccurate since list contains few movies with very less votes as well
correlation.sort_values('Correlation',ascending=False).head(10)
```

```python
#finding mean values of ratings
rating=pd.DataFrame(df.groupby('title')['rating'].mean())

rating.sort_values('rating',ascending=False).head()
```

```python
#finding the number of votes each movie got
rating['rating_count']=pd.DataFrame(df.groupby('title')['rating'].count())
rating.head()
```

```python
#adding rating count to correlation table
correlation = correlation.join(rating['rating_count'], rsuffix='_count')
correlation.head()
```

```python
rating.sort_values('rating_count',ascending=False).head()
```

```python
correlation.sort_values('Correlation',ascending=False).head(10)
```
